*    **Cadastro de alunos no sistema**




```python
#criado o banco de dados central da escola
escola_alunos = []
```

```python
continuar = True
while continuar == True:

  while True:
    nome = input('digite o nome completo do aluno: ').upper().strip()
    if nome != '':
      break
    else:
      print('o nome n pod ficar em branco, digite novamente')

  while True:
    data_de_nascimento = input('Digite a data de nascimento (DD/MM/AAAA): ').upper().strip().replace('/', '.')
    if len(data_de_nascimento) == 10 and data_de_nascimento.count('.') == 2:
      break
    else:
      print('digite a data no formato correto (DD/MM/AAAA)')

  while True:
    turma = input('digite a turma do aluno: ').upper().strip()
    if turma != '':
      break
    else:
      print('a turma n pod ficar em branco, digite novamente')

  aluno = {
      'nome' : nome,
      'data_de_nascimento' : data_de_nascimento,
      'turma' : turma
  }
  escola_alunos.append(aluno)
  while True:
    resposta = input('deseja continuar, digite s se sim e n se deseja encerrar').upper().strip()
    if resposta == 'S':
      break
    elif resposta == 'N':
      continuar = False
      break
    else:
      print('digite apenas s se deseja continuar ou n se deseja encerrar')
    continue


```

**criando provas pros alunos**

```python
for aluno in escola_alunos:
  print(f'digite as notas do aluno {aluno["nome"]}')

  while True:
    try:
      mat_1 = float(input('digite a primeira nota da materia matematica: '))
      break
    except ValueError:
      print('Entrada inválida. Por favor, digite um número para a nota.')
  while True:
    try:
      mat_2 = float(input('digite a segunda nota da materia matematica: '))
      break
    except ValueError:
      print('Entrada inválida. Por favor, digite um número para a nota.')

  while True:
    try:
      port_1 = float(input('digite a primeira nota da materia portugues: '))
      break
    except ValueError:
      print('Entrada inválida. Por favor, digite um número para a nota.')
  while True:
    try:
      port_2 = float(input('digite a segunda nota da materia portugues: '))
      break
    except ValueError:
      print('Entrada inválida. Por favor, digite um número para a nota.')

  while True:
    try:
      geo_1 = float(input('digite a primeira nota da materia geografia: '))
      break
    except ValueError:
      print('Entrada inválida. Por favor, digite um número para a nota.')
  while True:
    try:
      geo_2 = float(input('digite a segunda nota da materia geografia: '))
      break
    except ValueError:
      print('Entrada inválida. Por favor, digite um número para a nota.')

  notas = {
      'matematica' : [mat_1, mat_2],
      'portugues' : [port_1, port_2],
      'geografia' : [geo_1, geo_2]
  }
  aluno['notas'] = notas
```

```python
for aluno in escola_alunos:
  print(f'aluno: {aluno["nome"]}, {aluno["data_de_nascimento"]}, {aluno["turma"]}, {aluno["notas"]}')
```

pesquisando aluno pelo nome



```python
print('digite o nome do aluno q deseja pesquisar')
aluno_a_pesquisar = input().upper().strip()
aluno_encontrado = False

for aluno in escola_alunos:
  if aluno_a_pesquisar in aluno['nome']:
    aluno_encontrado = True

    print(f'aluno encontrado: {aluno["nome"]}')

    for chave, valor in aluno.items():
      chave_formatada = chave.upper()

      if chave_formatada == 'NOTAS':
        print(f'{chave_formatada}:')
        for materia, lista_notas in valor.items():
          print(f"  - {materia.capitalize()}: {lista_notas[0]} e {lista_notas[1]}")


      elif chave == 'data_de_nascimento':
        print(f'{chave_formatada}: {str(valor).replace(".", "/")}')

      else:
        print(f'{chave_formatada}: {valor}')

if not aluno_encontrado:
  print('aluno nao encontrado')
```

calcular media



```python
# 1. Definindo a função que gera o relatório de toda a escola
def gerar_relatorio_escola(lista_de_alunos):
    print('========================================')
    print('       RELATÓRIO GERAL DE MÉDIAS        ')
    print('========================================')

    # O laço passa por cada aluno da lista que você passar para a função
    for aluno in lista_de_alunos:
        if 'notas' in aluno:
            print(f"\nAluno: {aluno['nome']}")
            print("-" * 40)

            # Calcula a média usando sum() e len() para cada matéria
            for materia, lista_notas in aluno['notas'].items():
                media = sum(lista_notas) / len(lista_notas)
                print(f"  - {materia.upper()}: Média = {media:.1f}")

            print("-" * 40)

    print('\nFim do relatório!')

```

```python
gerar_relatorio_escola(escola_alunos)
```
